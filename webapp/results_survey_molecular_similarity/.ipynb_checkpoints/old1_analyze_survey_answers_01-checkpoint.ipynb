```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#The current notebook should be
#in the main "molecular similarity survey" directory.
dir_survey = Path.cwd()
```

```python
file_answers = Path(dir_survey, "all_answers_survey.csv")
df_answers = pd.read_csv(file_answers)
df_answers
```

```python
colname_score_2d = "tanimoto_cdk_Extended"
colname_score_3d = "TanimotoCombo"

colname_dist = "pchembl_distance"

colname_target = "target_name"

colname_subset = "pair_type"

colname_similar = "answer_similar"

colname_experience = "answer_experience"

colname_user = "id_user"

colname_pair = "id_surveyPair"
```

```python
categories_similar = ["Yes", "No"]
```

```python
categories_experience = ["PhD Student",
                         "Postdoc",
                         "Professor / Researcher",
                         "None of the above",
                         ]
```

```python
categories_subset = ["sim2D,sim3D",
                     "dis2D,sim3D",
                     "sim2D,dis3D",
                     "dis2D,dis3D",
                     ]
```

```python
targets_names = ["HERG",
                 "5HT2B",
                 "CYP2D6",
                 ]
```

```python
nicename_score_2d = (colname_score_2d
                     .replace("_", " ")
                     .title()
                     .replace("Cdk", "CDK")
                     )
nicename_score_3d = colname_score_3d

nicename_dist = "pChEMBL Distance"

nicename_target = "Target"

nicename_similar = "Similar"

nicename_experience = "Academic Qualification"
```

```python
FIGSIZE_GOLD = (9.556, 5.906) # Golden Rectangle / (in, in)

FONTSIZE = 14
```

```python
lim_score_2d = (0, 1)
lim_score_3d = (0, 2)
```

### Plot Number of Users who completed the survey by Academic Experience.

```python
#The original DataFrame has a row for each answer;
#each user was asked 5 questions, so blocks of 5 answers in original
#DataFrame correspond to a single user.
counts_experience = (df_answers
                     .groupby(colname_experience)[colname_user]
                     .nunique()
                     .rename("count")
                     .reset_index()
                     )

#Only those who completed the survey!
tot_users = counts_experience["count"].sum()


fig, ax = plt.subplots(constrained_layout=True, figsize=FIGSIZE_GOLD)
sns.barplot(data=counts_experience,
            x=colname_experience,
            y="count",
            order=categories_experience,
            )

ax.set_title(f"{tot_users} users completed the survey", fontsize=FONTSIZE + 3)
ax.set_ylabel("Number of Users", fontsize=FONTSIZE)
ax.set_xlabel(nicename_experience, fontsize=FONTSIZE);
```

### Plot Answer Percentages in the different calculated similarity subsets.
This is technically a "stacked normalized countplot". Creating it is more complex than it seems!
Seaborn does not allow stacking or normalizing. Dexplot seems ideal, but is no longer actively developed, it raises warnings, and ordering the columns does not work!

My plot is based on [this](https://stackoverflow.com/questions/34615854/seaborn-countplot-with-normalized-y-axis-per-group) and [this](https://stackoverflow.com/questions/50319614/count-plot-with-stacked-bars-per-hue) stackoverflow answers.

```python
fracs_similar = (df_answers
                 .groupby(colname_subset)[colname_similar]
                 .value_counts(normalize=True)
                 #Rename the Series so that using `reset_index` is possible.
                 .rename("fraction")
                 #Convert multi-index Series to DataFrame.
                 .reset_index()
                 )
fracs_similar
```

```python
percent_similar = (fracs_similar
                   #Transform to format more suitable to create stacked barplot.
                   .pivot(columns=colname_similar,
                          index=colname_subset,
                          #Specify `values` to avoid hierarchically indexed columns.
                          values="fraction")
                   #Convert fraction to percentage.
                   .mul(100)
                   #Sort index and columns in the order in which they will appear in barplot.
                   .loc[categories_subset[::-1], categories_similar]
                   )


fig, ax = plt.subplots(constrained_layout=True, figsize=FIGSIZE_GOLD)
percent_similar.plot(kind="barh", stacked=True, ax=ax)

ax.legend(title=nicename_similar)

ax.set_xlim(0, 100)
ax.set_xlabel("Answer Percentage", fontsize=FONTSIZE)
ax.set_ylabel("Pair Type", fontsize=FONTSIZE);
```

```python
counts_pairs = (df_answers
                .groupby(colname_subset)[colname_pair]
                .value_counts()
                .rename("n_answers")
                .reset_index()
                )
counts_pairs
```
