## Define variables
Contained in the input data, and necessary to create nice figures.

```python
colname_score_2d = "tanimoto_cdk_Extended"
colname_score_3d = "TanimotoCombo"

colname_dist = "pchembl_distance"

colname_target = "target_name"

colname_subset = "pair_type"

colname_pair = "id_surveyPair"

colname_n_ans = "n_answers"

colname_n_simil = "n_similar"

colname_frac_simil = "frac_similar"
```

```python
categories_subset = ["dis2D,dis3D",
                     "dis2D,sim3D",
                     "sim2D,dis3D",
                     "sim2D,sim3D",
                     ]
```

```python
targets_names = ["HERG",
                 "5HT2B",
                 "CYP2D6",
                 ]
```

```python
nicename_score_2d = (colname_score_2d
                     .replace("_", " ")
                     .title()
                     .replace("Cdk", "CDK")
                     )
nicename_score_3d = colname_score_3d

nicename_dist = "pChEMBL Distance"

nicename_target = "Target"

nicename_subset = "Pair Type"

nicename_similar = "Similar"

nicename_experience = "Academic Qualification"

nicename_n_ans = "Number of Answers"

nicename_ans_percent = "Answer Percentage"

nicename_simil_percent = "Similarity Percentage"

nicename_pair = "Pair ID"
```

```python
lim_score_2d = (0, 1)
lim_score_3d = (0, 2)
```

### Plot answer percentages for each similarity category
We will use the actual percentages in each category, **not** the mean of the percentages of every pair in each category!\
The results will be slightly different!

So, we will use: $\frac{N_{Yes}}{N_{Tot}} \times 100$ in each category!

```python
fracs_subsets = (gby_subset
                 .apply(lambda x: x[colname_n_simil].sum() / x[colname_n_ans].sum())
                 .rename("Yes")
                 .to_frame()
                 #Sort index in the order in which it will appear in barplot.
                 .loc[categories_subset, :]
                 )
fracs_subsets["No"] = 1 - fracs_subsets["Yes"]
fracs_subsets.round(3)
```

```python
fig, ax = plt.subplots(**kwargs_fig_basic)
fracs_subsets.mul(100).plot(kind="barh", stacked=True, ax=ax)

ax.set_xlim(0, 100)
ax.set_xlabel(nicename_ans_percent, fontsize=FONTSIZE)
ax.set_ylabel(nicename_subset, fontsize=FONTSIZE)

file_fig = Path(dir_out, "similarity_percent_categories.pdf")
fig.savefig(file_fig)
```

Also add a regression line, and calculate regression parameters.

```python
fig, ax = plt.subplots(**kwargs_fig_basic)
sns.scatterplot(data=df_nice,
                x=nicename_simil_percent,
                y=nicename_dist,
                hue=nicename_subset,
                style=nicename_subset,
                hue_order=reversed(categories_subset),
                style_order=reversed(categories_subset),
                )
#Get X and Y axis limits used by `scatterplot`,
#since `regplot` will change them.
xlim, ylim = ax.get_xlim(), ax.get_ylim()

sns.regplot(data=df_nice,
            x=nicename_simil_percent,
            y=nicename_dist,
            color="black",
            scatter_kws={"s": 0},
            )
#Reset X and Y axis limits that were used by `scatterplot`.
ax.set_xlim(xlim)
ax.set_ylim(ylim)

formula_regression = f"{colname_dist} ~ {colname_frac_simil}"
results_regression = smf.ols(formula_regression, df_agg).fit()
ax.set_title(f"R$^2$ = {results_regression.rsquared:0.2f}")
print(results_regression.summary())

ax.xaxis.label.set(fontsize=FONTSIZE)
ax.yaxis.label.set(fontsize=FONTSIZE)

fig_file = Path(dir_out, "correlation_pChEMBL_similarity.pdf")
fig.savefig(fig_file)
```

### Pairplot with KDEs and Correlation Coefficients
The code to create this complex pairplot was taken [here](https://towardsdatascience.com/visualizing-data-with-pair-plots-in-python-f228cf529166).

```python
# Function to calculate correlation coefficient between two arrays
def corr(x, y, **kwargs):
    
    # Calculate the value
    coef = np.corrcoef(x, y)[0][1]
    # Make the label
    label = r'$\rho$ = ' + str(round(coef, 2))
    
    # Add the label to the plot
    ax = plt.gca()
    ax.annotate(label, xy = (0.2, 0.95), size=20, xycoords=ax.transAxes)
    
# Create a pair grid instance
grid = sns.PairGrid(data= df_nice)

# Map the plots to the locations
grid = grid.map_upper(plt.scatter)
grid = grid.map_upper(corr)
grid = grid.map_lower(sns.kdeplot)
grid = grid.map_diag(plt.hist, edgecolor="black", bins=10)

fig = plt.gcf()
fig.set_size_inches(FIGSIZE_SQUARE)

file_fig = Path(dir_out, "pairplot.pdf")
fig.savefig(file_fig)
```

### Box Plot of Similarity Vote Percentages in each category
Same info but more complete and nuanced than the "Yes / No Averages in each category" plot.

```python
fig, ax = plt.subplots(**kwargs_fig_basic)
sns.boxplot(data=df_nice,
            x=nicename_simil_percent,
            y=nicename_subset,
            #(!)the `categories_subset` order is here reversed because Seaborn, when given an order,
            #uses an order which is opposite than the one used by Pandas plotting!
            #I want this plot to be in the same order as previous plots made with Pandas.
            order=categories_subset[::-1],
            )
sns.swarmplot(data=df_nice,
            x=nicename_simil_percent,
            y=nicename_subset,
            order=categories_subset[::-1],
            color=".25"
            )

ax.xaxis.label.set(fontsize=FONTSIZE)
ax.yaxis.label.set(fontsize=FONTSIZE)

file_fig = Path(dir_out, "boxplot_similarity_percent_categories.pdf")
fig.savefig(file_fig)
```

***Notes***:
* Both `sim2D,dis3D` and `dis2D,sim3D` boxes are centered in the 40-60% range.
* The `sim2D,dis3D` box is much narrower than the `dis2D,sim3D` box $\Rightarrow$ humans have stronger agreement in pairs of molecules that are similar in 2D, since humans are probably watching more the 2D visualizations than the 3D interactive representations, when making a decision!
