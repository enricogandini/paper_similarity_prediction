## Create Pipelines for single-feature models

```python
kwargs_logreg = {"random_state": 1,
                 "fit_intercept": True,
                 #(!)IMPORTANT! By default, L2 regularization is applied,
                 #but it is meaningless when using a single input feature!
                 #And we will be using only one Tanimoto column as input feature!
                 #Setting `penalty` to "none" drastically improves results!!
                 "penalty": "none",
                 "max_iter": 1000,
                 }


initial_pipelines = {}
for colname_feature in colnames_features:
    #Select a single column in a pipeline,
    #see https://stackoverflow.com/questions/62416223/how-to-select-only-few-columns-in-scikit-learn-column-selector-pipeline
    #`colname_feature` is in a list so output of transformer is a 2D array with just one column.
    selector_feature = ColumnTransformer([("selector", "passthrough", [colname_feature])],
                                         remainder="drop",
                                         )
    logreg = LogisticRegression(**kwargs_logreg)
    pipeline = make_pipeline(selector_feature, logreg)
    initial_pipelines[colname_feature] = pipeline
```

## Fit single-feature pipelines

```python
fitted_pipelines = {}
for used_feature, pipeline in initial_pipelines.items():
    pipeline.fit(X_train, y_train)
    fitted_pipelines[used_feature] = pipeline
```

## Evaluate Models on training-set
The simplest evaluations are done on the whole training-set, no test-set or validation-set.

```python
df_evals = [] #will contain results of all scores for all models
for used_feature, pipeline in fitted_pipelines.items():
    pred_proba = pipeline.predict_proba(X_train)[:, 1]
    df_agg[f"proba_{used_feature}"] = pred_proba
    pred = pipeline.predict(X_train)
    df_agg[f"pred_{used_feature}"] = pred
    
    correct = (y_train == pred).astype(int)
    n_correct = correct.sum()
    print(f"Model based on `{used_feature}`"
          f"correctly predicted {n_correct}/{n_train} molecular pairs.")
    df_agg[f"correct_{used_feature}"] = correct
    
    tmp = {"model": used_feature,
           "n_correct": n_correct,
           }
    metrics_dict = mlh.various_metrics_binary_classification(model=pipeline,
                                                             metrics=mlh.METRICS_BINARY_PROBA,
                                                             X=X_train,
                                                             y_true=y_train,
                                                             )
    tmp.update(metrics_dict)
    df_evals.append(tmp)


df_evals = pd.DataFrame(df_evals)
df_evals
```

```python
file_evals = Path(dir_models_analysis, "metrics_single_feature_models.csv")
df_evals.to_csv(file_evals, index=False)
```

Use XOR operator to get molecular pairs that were correctly predicted by any one of the two models, but not by both models\

***Notes***:
* `sim2D,sim3D` and `dis2D,dis3D` sets were 100% correctly predicted by the two single-feature models.
* `sim2D,dis3D` set received 72% correct predictions by the 2D-only model, but only 32% correct predictions by the 3D-only model $\Rightarrow$ users only watched the 2D depictions (even though they had the 3D representations), so the 2D-only model reproduced the user judgements more closely.
* `dis2D,sim3D` set received ~50% correct predictions by the two single-feature models $\Rightarrow$ users were very unsure about molecular similarity in this set.

So, we can say that users watch more closely the 2D depictions than the 3D representations, and for this reason the 2D-only model is better in reproducing user judgements.\
This is also confirmed by the "answer percentage in each similarity category" graph (in the "analyze" notebook): `sim2D,dis3D` set received 56% of "Yes" answers, compared to only 51% of `dis2D,sim3D` set $\Rightarrow$ users gave an higher percentage of "Yes" votes to ambiguous molecules that are similar in 2D than to ambiguous molecules that are similar in 3D.

Also watch the "BoxPlot of Similarity Vote Percentages" graph (and related notes) that are in good agreement with these observations.

### Plot logistic curve of single-feature models

Create a DataFrame with `linspace` of the two scores. This is necessary because the pipelines take a DataFrame as input and select one column.

```python
X_linspace = {}
for colname, lim in lims_scores.items():
    X_linspace[colname] = np.linspace(*lim, 100)


X_linspace = pd.DataFrame(X_linspace)
```

```python
for colname_feature, pipeline in fitted_pipelines.items():
    nicename_feature = nicenames_scores[colname_feature]
    
    proba_linspace = pipeline.predict_proba(X_linspace)[:, 1]
    proba_linspace *= 100 # from frac to percent
    
    fig, ax = plt.subplots(**kwargs_fig_basic)
    #Plot feature and similarity judgements of original molecular pairs.
    ax.plot(df_agg[colname_feature],
            df_agg[colname_frac_simil] * 100,
            marker='o',
            linestyle='none',
            label="True"
            )
    #Plot logistic curve in linspace of feature.
    ax.plot(X_linspace[colname_feature], proba_linspace, label="Predicted")
    
    ax.legend()
    ax.set_xlabel(nicename_feature, fontsize=FONTSIZE)
    ax.set_ylabel(nicename_simil_percent, fontsize=FONTSIZE)
    
    file_fig = Path(dir_models_analysis, f"logreg_curve_single_feature_{colname_feature}.pdf")
    fig.savefig(file_fig)
    plt.show()
```
