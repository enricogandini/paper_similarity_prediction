## Define Training set

```python
colnames_features = [colname_score_2d, colname_score_3d]

X_train = df_agg.loc[:, colnames_features].copy()
n_train = X_train.shape[0]
```

Training set `y` labels will also be added to the original DataFrame, for easier analyses.

```python
human_similarity = df_agg[colname_frac_simil] # fraction of human experts that considered a pair of molecules to be similar

thres_low = 0.4
thres_high = 0.6
#(!)The last break is `1.0001` instead of `1.0` since
#I need non-overlapping intervals
#(so, `closed="both"` option in `from_breaks` method is not possible),
#and I need to include the right extreme `1.0`
#(otherwise, pairs with `frac_similar` equals `1.0` would be in a NaN bin).
breaks = [0, thres_low, thres_high, 1.00001]
bins_classes = pd.IntervalIndex.from_breaks(breaks,
                                            closed="left",
                                            )
labels_bins_classes = ["Not Similar", "Uncertain", "Similar"]

y_train = pd.cut(human_similarity,
                 bins=bins_classes,
                 )
y_train = y_train.cat.rename_categories(labels_bins_classes)

colname_judged = "judged_similar"
y_train.name = colname_judged
df_agg[colname_judged] = y_train

#Check that all items in `human_similarity` were assigned to a bin.
if y_train.isna().any():
    raise ValueError("NaN values in `y_train`! Check the binning!")

y_train
```

Counts of new multi-class labelling.

```python
counts_y_train = (pd.value_counts(y_train)
                  .rename("train_labels")
                  .loc[labels_bins_classes]
                  .to_frame()
                  )
counts_y_train
```

## Create Pipelines for single-feature models

```python
class_weight = "balanced" #Balance an unbalanced multi-class dataset.
kwargs_logreg = {"random_state": 1,
                 "fit_intercept": True,
                 #(!)IMPORTANT! By default, L2 regularization is applied,
                 #but it is meaningless when using a single input feature!
                 #And we will be using only one Tanimoto column as input feature!
                 #Setting `penalty` to "none" drastically improves results!!
                 "penalty": "none",
                 "max_iter": 1000,
                 "class_weight": class_weight,
                 }


initial_pipelines = {}
for colname_feature in colnames_features:
    #Select a single column in a pipeline,
    #see https://stackoverflow.com/questions/62416223/how-to-select-only-few-columns-in-scikit-learn-column-selector-pipeline
    #`colname_feature` is in a list so output of transformer is a 2D array with just one column.
    selector_feature = ColumnTransformer([("selector", "passthrough", [colname_feature])],
                                         remainder="drop",
                                         )
    logreg = LogisticRegression(**kwargs_logreg)
    pipeline = make_pipeline(selector_feature, logreg)
    initial_pipelines[colname_feature] = pipeline
```

## Fit single-feature pipelines

```python
fitted_pipelines = {}
for used_feature, pipeline in initial_pipelines.items():
    pipeline.fit(X_train, y_train)
    fitted_pipelines[used_feature] = pipeline
```

```python
pip = fitted_pipelines["tanimoto_cdk_Extended"]
lr = pip.named_steps["logisticregression"]
pred = pip.predict(X_train)
pred_proba = pip.predict_proba(X_train)
```

## Evaluate Models on training-set
The simplest evaluations are done on the whole training-set, no test-set or validation-set.

```python
df_evals = [] #will contain results of all scores for all models
merged_counts_pred = []
df_fracs_correct = []
for used_feature, pipeline in fitted_pipelines.items():
    pred = pipeline.predict(X_train)
    df_agg[f"pred_{used_feature}"] = pred
    
    #Count classes predicted by the model.
    counts_pred = (pd.value_counts(pred)
                   .rename(used_feature)
                   .loc[labels_bins_classes]
                   )
    merged_counts_pred.append(counts_pred)
    
    #Fractions of correct predictions
    fracs_correct = mlh.fracs_correct_predictions(pipeline,
                                                  X_train,
                                                  y_train,
                                                  )
    fracs_correct = (fracs_correct
                     .loc[labels_bins_classes]
                     .rename(used_feature)
                     )
    df_fracs_correct.append(fracs_correct)
    
    
    correct = (y_train == pred).astype(int)
    n_correct = correct.sum()
    rand_index = rand_score(y_train, pred)
    print(f"Model based on `{used_feature}`"
          f"correctly predicted {n_correct}/{n_train} molecular pairs.")
    df_agg[f"correct_{used_feature}"] = correct
    
    tmp = {"model": used_feature,
           "n_correct": n_correct,
           "rand_index": rand_index,
           }
    metrics_dict = mlh.various_metrics_multi_classification(model=pipeline,
                                                            metrics=mlh.METRICS_MULTI_PROBA,
                                                            X=X_train,
                                                            y_true=y_train,
                                                            )
    tmp.update(metrics_dict)
    df_evals.append(tmp)


merged_counts_pred = pd.DataFrame(merged_counts_pred).T
df_fracs_correct = pd.DataFrame(df_fracs_correct).T
df_evals = pd.DataFrame(df_evals)
df_evals
```

### Correct predictions by single-feature models

```python
tot_correct_subsets = df_agg.pivot_table(index=colname_subset,
                                         values=[f"correct_{colname_score_2d}", f"correct_{colname_score_3d}"],
                                         aggfunc="sum",
                                         )

#Make pivoted DataFrame nicer for graph.
tot_correct_subsets = (tot_correct_subsets
                       #Nice names to columns.
                       .rename(columns={f"correct_{colname_score_2d}": nicename_score_2d,
                                        f"correct_{colname_score_3d}": nicename_score_3d})
                       #Reorder rows and columns.
                       .loc[categories_subset, [nicename_score_2d, nicename_score_3d]]
                       )
```

```python
frac_correct_subsets = tot_correct_subsets / n_pairs_each_subset

fig, ax = plt.subplots(**kwargs_fig_basic)
(frac_correct_subsets * 100).plot(kind="barh", ax=ax)

ax.set_title("Correct predictions by single-feature models",
             fontsize=FONTSIZE,
             )
ax.set_xlabel(nicename_percent_pairs_subset, fontsize=FONTSIZE)
ax.set_ylabel(nicename_subset, fontsize=FONTSIZE)
ax.set_xlim(lim_percent)

file_correct_subsets = Path(dir_models_analysis, "correct_predictions_per_subsets_by_single_feature_models.pdf")
fig.savefig(file_correct_subsets)
```
